```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
url = 'https://raw.githubusercontent.com/Amaniiitd/Fetal-Health-classification/main/Dataset/fetal_health.csv'
df = pd.read_csv(url)
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
print(df.fetal_health.unique())
plt.figure(figsize=(8,5))
sb.countplot(data= df, x="fetal_health")
```

```python
plt.figure(figsize=(20, 10))
dataplot = sb.heatmap(df.corr(), cmap='PiYG', annot=True)
  
# displaying heatmap
plt.show()
```

```python
df.corr().abs().fetal_health.sort_values(ascending=False)
```

```python
plt.figure(figsize=(20, 10))
X=df.drop(["fetal_health"],axis=1)
y=df["fetal_health"]
sb.boxenplot(data = X)
plt.xticks(rotation=90)
plt.show()
```

```python
def feature_normalize(X):
    #calculating the mean
    mu = np.mean(X, axis = 0)  
    
    #calculating the standard deviation
    sigma = np.std(X, axis= 0, ddof = 1)  # Standard deviation (can also use range)
    
    #Normalising the dataset
    X_norm = (X - mu)/sigma

    return X_norm
X=feature_normalize(X)

plt.figure(figsize=(20, 10))
sb.boxenplot(data = X)
plt.xticks(rotation=90)
plt.show()
```
